A coding exercise on trasforming data from wide to long format and back again.  Git repo created and pushed.

```python
# push test
import pandas as pd
import plotnine as p9
```

```python
orig = pd.read_csv(".\data\wl_table.csv")
```

```python
orig
```

```python
column_list = orig.columns[3:9]
id_list = ['Name', 'Dept', 'Project']
```

```python
long = pd.melt( orig, id_vars = id_list , value_vars = column_list, var_name = 'Week', value_name = 'Sales' )
```

```python
long
```

Reshape that long format table back to the original wide format.

```python
wide = long.pivot_table( index = id_list, columns = 'Week', values = 'Sales' )
```

```python
wide
```

The columns are now in alpha order instead of date order.  In case the column index is being used elsewhere, this should be fixed.

```python
wide_rc = wide.reindex(columns = column_list)
```

```python
wide_rc
```

```python
wide_rc.to_csv(".\data\wide_rc.csv", header = True)
```

So the file on disk is the same as the original except that the records (rows) are in alpha order instead of political order.  That could be fixed, but AFAIK would not be a problem in other parts of related code.
